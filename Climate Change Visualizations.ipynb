```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt


file = 'https://course-resources.minerva.kgi.edu/uploaded_files/mke/rx9A0n/s1-student-version--ea-assignment--climate-change--land-ice-mass-data---sheet1.csv'
icedata = pd.read_csv(file, skiprows=10)


fig, ax = plt.subplots()
timedata = list(icedata["TIME (year.decimal fraction)"].values)
greenlandmass = list(icedata["Greenland mass (Gt)"].values)
antarc = list(icedata["Antarctica mass (Gt)"].values)

#Line graph
ax.plot(timedata,greenlandmass, color="red", label="Greenland")
ax.plot(timedata,antarc, label="Antartica")
ax.set(xlabel = 'Year\n Fig.4 Amount of Melting Ice in Greenland and Antarctica\n  Graph of mass anomaly of ice melting in Antartica is more compressed than Greenland.\n As from 2009 -14, ice melting anomaly of greenland is about 100% more than in antartica.', ylabel = 'Mass Anomaly of Ice Melting (Gt)')
ax.grid()
plt.legend()
plt.show()

#Scatter plot
fig, ax = plt.subplots()
ax.scatter(greenlandmass, antarc, s=100, color="red", label="Greenland vs Antartica")
plt.plot(np.unique(greenlandmass), np.poly1d(np.polyfit(greenlandmass, antarc, 1))(np.unique(greenlandmass)))
print(np.corrcoef(greenlandmass, antarc))
ax.set(xlabel = 'Greenland Mass of Melting Ice (Gt)\n Fig.2 Amount of Melting Ice in Greenland and Antarctica Over Time.\n Strong positive linear correlation between Melting of Ice in Green land\n and Antartica.\n Correlation can be estimated to be around 0.94', ylabel = 'Antartica Mass of Melting Ice (Gt)')
ax.grid()
plt.legend()
plt.show()

#Histogram
fig, ax = plt.subplots()
ax.hist(antarc,label="Antartica", edgecolor="black", alpha=0.7)
ax.hist(greenlandmass,color="red", label="Greenland", edgecolor="black", alpha=0.5)
ax.set(xlabel = 'Amount of Icemelt (Gt)', ylabel = 'Frequency', title = 'Fig.3 Amount of Melting Ice in Greenland and Antarctica\n Greenland has more outliers hence a bigger average amount of\n icemelt while antartica has less outliers hence greater median\n value of amount of icelmelt. However, antartica records the\n highest relative frequency of anomaly icemelt of about375- 500Gt\n compared to Greenland anomaly of about 0-125 Gt.')
plt.legend()
plt.show()
```

```python
#barcharts
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt


file = 'https://course-resources.minerva.kgi.edu/uploaded_files/mke/rx9A0n/s1-student-version--ea-assignment--climate-change--land-ice-mass-data---sheet1.csv'
icedata = pd.read_csv(file, skiprows=10)


timedata = list(icedata["TIME (year.decimal fraction)"].values)
greenlandmass = list(icedata["Greenland mass (Gt)"].values)
antarc = list(icedata["Antarctica mass (Gt)"].values)

g1=np.mean(greenlandmass[:35])
g2=np.mean(greenlandmass[35:73])
g3=np.mean(greenlandmass[73:108])
g4=np.mean(greenlandmass[108:140])
a1=np.mean(antarc[:35])
a2=np.mean(antarc[35:73])
a3=np.mean(antarc[73:108])
a4=np.mean(antarc[108:140])
n_groups = 4
means_g = (g1, g2, g3, g4)
stdg=np.std(means_g)
means_a = (a1, a2, a3, a4)
stda=np.std(means_a)
# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.7
error=[stdg, stda]

rects1 = plt.bar(index, means_g, bar_width, bottom=means_a, alpha=opacity,
color='red',
label='Greenland')

rects2 = plt.bar(index, means_a, bar_width, alpha=opacity,
color='blue',
label='Antartica')

ax.set(xlabel = 'Time Categories (years)', ylabel = 'Mean of Anomalies of Mass of Melting Ice (Gt)', title = 'Fig.4 Mean of Anomalies of Mass of Melting Ice in Different year periods\n In 3 year intervals the postive and negative mean anomalies\n for Greenland is higher than in Antartica\n this tells you the extent of spread of mean anomalies\n and how wide the anomaly difference\n in Greenland is')
plt.xticks(index, ('2002-2005', '2005-2008', '2008-2011', '2011-2014'))
plt.legend()
ax.grid()
plt.rcParams["figure.figsize"] = [16,9]

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt


file = 'https://course-resources.minerva.kgi.edu/uploaded_files/mke/nbpl5V/s1--ea--signature-2--scientific-method--vostok-data-with-temp-calculated.xlsx---vostok-data--1-.csv'
icedata = pd.read_csv(file)


fig, ax = plt.subplots()
depth = list(icedata["depth"].values)
rdepth= depth[1:]
rdepth = [int(i) for i in rdepth]
iage = list(icedata["ice_age"].values)
riage= iage[1:]
riage = [float(i) for i in riage]
gage = list(icedata["gas_age"].values)
rgage=gage[1:]
rgage = [float(i) for i in rgage]
agediff=list(np.array(riage) - np.array(rgage))
co2 = list(icedata["CO2"].values)
rco2=co2[1:]
rco2 = [float(i) for i in rco2]
temp = list(icedata["Temperature (calculated)"].values)
rtemp=temp[1:]
rtemp = [float(i) for i in rtemp]

#Line graph
ax.plot(rdepth,riage, color="red", label="Ice Age")
ax.plot(rdepth,rgage, color="blue", label="Gas Age")
ax.set(xlabel = 'Ice depth (m)\n Fig.1 Ice and Gas Ages and corresponding ice depth \n  At same depth ice age is relatively greater than gas age.\n As depth of ice deepens from 1000 to 2000 metres,\n both gas and ice ages increase a little over 2 folds.', ylabel = 'Age (ka)')
ax.grid()
plt.legend()
plt.show()

#Scatterplot
fig, ax = plt.subplots()
ax.scatter(rtemp, agediff, color="red", label="Age difference vs. Temperature")
plt.plot(np.unique(rtemp), np.poly1d(np.polyfit(rtemp, agediff, 1))(np.unique(rtemp)))
print(np.corrcoef(rtemp, agediff))
ax.set(xlabel = 'Temperature calculated (Celsius)\n Fig.2 Correlation between age difference in CO2 and ice and temperature of ice\n  Strong negative linear correlation between age difference and temperature\n Correlation can be estimated to be around -0.89', ylabel = 'Age difference (ka)')
ax.grid()
plt.legend()
plt.show()

#Scatter plot
fig, ax = plt.subplots()
ax.scatter(rco2, rtemp, color="red", label="Temperature vs. Carbon dioxide")
plt.plot(np.unique(rco2), np.poly1d(np.polyfit(rco2, rtemp, 1))(np.unique(rco2)))
print(np.corrcoef(rco2, rtemp))
ax.set(xlabel = 'Carbon dioxide (ppmv)\n Fig.3 Correlation between amount of CO2 in atmosphere and temperature.\n Strong positive linear correlation between CO2 content and calculated temperature.\n Correlation can be estimated to be around 0.81', ylabel = 'Temperature calculated (Celsius)')
ax.grid()
plt.legend()
plt.show()
```
