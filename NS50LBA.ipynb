```python
import pandas as pd
file = 'nsfloods.csv'
htf = pd.read_csv(file)
htf
```

```python
yr_past=list(htf["yearspast"])
flood_hist=list(htf["floodshist"])

import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(13,6.5))
plt.bar(yr_past, flood_hist, width=0.8, color="red", alpha=0.7)
plt.title("")
plt.xlabel("Year")
plt.ylabel("Number of days per year with occurrences of High Tide Flooding")
plt.xticks(np.arange(1920, 2030, step=10))
plt.grid()
plt.show()
print("Fig.1 Number of days with occurrences of annual High Tide Flooding along San Francisco Bay from 1920 to 2016.\nOnce sea level exceeds a treshold of 0.57m in San Francisco, High Tide Flooding is bound to occur.\nRetrieved from https://tidesandcurrents.noaa.gov ")

```

```python
yr_predicted=list(htf["yearspredicted"])
low=list(htf["lowscenario"])
int_low=list(htf["int_low"])
intermediate=list(htf["int"])
int_high=list(htf["int_high"])
high=list(htf["high"])
extreme=list(htf["extreme"])
y=[]
for i in yr_predicted:
    y=y+[365]
    

plt.figure(figsize=(13,6.5))
plt.plot(yr_predicted, low, label="Low scenario")
plt.plot(yr_predicted, int_low, label="Intermediate-Low scenario")
plt.plot(yr_predicted, intermediate, label="Intermediate scenario")
plt.plot(yr_predicted, int_high, label="Intermediate-High scenario")
plt.plot(yr_predicted, high, label="High scenario")
plt.plot(yr_predicted, extreme, label="Extreme scenario")
plt.plot(yr_predicted, y, color="black", linestyle="dashed", label="Maximum number of days a year")
plt.xticks(np.arange(2000, 2110, step=10))
plt.xlabel("Year")
plt.ylabel("Number of days per year with occurences of High Tide Flooding")
plt.legend()
print("Fig.2 Projected number of days with occurences of High Tide Flooding per year along the coastline of San Francisco for\n6 scenario types(low, intermediate low, intermediate, intermediate high, high and extreme) from the years 2000 to 2100\n. High Tide Flooding in assessed by the exceedance at or above the treshold of 0.57m The scenarios assessed are the global rise scenarios of the US Federal Interagency Sea Level Rise and Coastal Flood Hazard Task.\n Retrieved from https://tidesandcurrents.noaa.gov")
```
