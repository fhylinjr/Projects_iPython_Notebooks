#### 3.1.2 Compute descriptive statistics

Using Python, calculate the mean, median, mode, range, and standard deviation of your variable. Print these values. If you use a library function, you need to explain how it works with detailed comments. Do not blindly use library functions! 


```python
homeless=[41, 8, 11, 12, 39, 20, 11, 1, 0, 0, 11, 12]

#sorting list in ascending order
shomeless=homeless #duplicating the homeless list to sort the new list but keep the old list in order of increasing block number.
for i in range (len(shomeless)):#loop through the list from first to last position of list
    for j in range(i + 1, len(shomeless)):#loop through the list from second to last postion of list
        if(shomeless[i] > shomeless[j]):#checks for the numerial size of any 2 adjacent items in a list and sees if the item on left is bigger than on right
            temp = shomeless[i] #if left item is bigger, stores it under a new variable called temp
            shomeless[i] = shomeless[j]#then swaps left item with right item
            shomeless[j] = temp#finally right item takes the left item/value previously stored under temp
print("Arranged homeless list for 12 blocks: ",shomeless)#prints updated list

#calculating mean
total=0.0 #accumulator variable
for i in shomeless:#loops through sorted list
    total+=i #each item in list is added to total till loop is complete.
mean= total/len(shomeless)#total divided by number of items in list.
print("Mean number of homeless people: ",mean, "people")#prints mean

#calculating median
if len(shomeless)%2 == 1:#checks if total number of items is odd
    median = shomeless[len(shomeless)//2]#if so then median will be the value directly in the middle of the list which is found using integer division
elif len(shomeless)%2 == 0:#checks if total number of items is even
    median = (shomeless[len(shomeless)//2] + shomeless[(len(shomeless)//2)-1])/2 #if so median will be arithmetic mean of 2 values in middle of list found by using integer division 
print("Median number of homeless people: ",median, "people")#prints median

#calculating mode
itr=1 #sets highest number of count of any item in the list to 1
for i in shomeless:#loops through list
    if shomeless.count(i)>itr: #checks if the tally of any item in list is more than highest number recorded in variable.The .count() method returns the number of occurrences of an element in a list.
        mode=i# if so then sets the value in the list to variable mode anytime condition is satisfied 
        itr+=1# increases highest times a number has occured 
    else: 
        itr=itr #checks if no item in list satisfies initial condition
print("Mode is: ",mode, "people")#prints mode

#calculating range
print ("Range: ", shomeless[-1]-shomeless[0])#finds difference between values of last and first position in sorted list and prints the result.

#calculating standard deviation
dev=0.0#accumulator variable
for i in shomeless:
    dev=dev+((i-mean)**2)#each item is subtracted from mean which is squared and added to accumulator variable.
sdev=(dev/len(shomeless))**0.5 #sets root mean square calculation to variable called sdev.
print("Standard deviation: ",sdev)#prints standard deviation


```

## Part 4: Probability Considerations [#probability, #algorithms, #simulation]

Suppose something unfortunate happened: you stole too many napkins for your Fermi estimate, so you decided to write all of your variable measurements on separate napkins, one napkin for each block. On your way back to res, the wind picked up and blew them all away! Luckily, you managed to collect all of the napkins, but now the data is totally randomly reordered, meaning that you have no idea which napkin corresponds to which block. Suppose that you tried to just guess randomly which napkin goes with which block. In other words, you randomly assign each napkin to a given block.

### 4.1 Matching Napkins Simulation

What is the probability that you are unlucky, and sadly NONE of the napkins are matched to the correct block (you guessed all of them wrong)? Write a simulation in Python to estimate this probability. Ensure that your code has sufficiently detailed comments to explain the key steps in the solution. 

```python
from random import shuffle #imports shuffle method from random library
napkin_list=[] #sets list for storing napkin numbers
block_list=[] #sets list for storing block numbers
no_matches=0 #variable for storing number of instanaces when you were unlucky
num_sim=1000000 #number of simulations variable
for i in range(1,13): #iterates through numbers 1 to (13-1) or 12
    napkin_list.append(int(i)) #adds the iterated numbers to napkin list
    block_list.append(int(i)) #adds the iterated numbers to block list
for i in range(num_sim): #iterates for total number of simulations
    matches=0 #each time a new simulation is run sets matches varible to 0
    random.shuffle(napkin_list) # each simulation requires items in napkin to be shuffled
    for i in range(12): #loops through the items in napkin list
        if napkin_list[i]==block_list[i]: #checks if value of napkin equals value of of block at same position
            matches+=1 #increases number of matches by 1 every time a block and napkin number correspond at same position
    if matches==0:# checks if number of matches for a particular simulation was 0
        no_matches+=1 #if so increases number of no matches by 1 every time a simulation completes and the condition is satisfied
prob_unlucky=(no_matches/num_sim) #calculates probability as total number of simulations with no_matches counted over total number of simulations counted
print("Probability of being unlucky: ",prob_unlucky) #prints results
```

DETAILED EXPLANATION OF SIMULATION

My simulation initially creates a list of block and napkin numbers from 1-12. Then it sets a "no_match" variable for storing the number of times for any simulation you were unlucky. After it sets your desired amount of simulations which should be a huge number to achieve the frequentist approach in estimating the probability. For each of the simulations,  napkins must be shuffled and compared with the blocks list in terms of values and corresponding positions. If the values and their positions correspond, then for that simulation, there was a match. If not, then you were unlucky for the simulation, and it is recorded and accumulated in the "no match" variable. This is repeated for the entire number of simulations, and probability is calculated as the number of unlucky simulations over total number of simulations. 
    This simulation is different from analytic calculations because I was able to answer the question of unluckiness even though the model was too complicated to solve. With analytics, I may have to use theorems, assumptions, and principles such as the Inclusion-Exclusion principle or Mathematical induction, which involves a lot of complex calculations. 
    Also, I realized as the number of simulations was small about less than 3, the result was decreasing towards 0.3, but even as n becomes so large, it converges at about 0.36. Consequently, my result can be interpreted using the Law of Large Numbers, which states that As more observations are collected, the proportion of occurrences with a particular outcome converges to the probability of that outcome.


### 4.3 *Optional*: Expected Value

What is the expected number of napkins that will be correctly matched to the corresponding block? Estimate this probability using a simulation and provide a detailed interpretation of the result.

```python
from random import shuffle
napkin_list=[]
block_list=[]
fmatches=0 #accumulator variable of total number of matches
num_sim=1000000
for i in range(1,13):
    napkin_list.append(int(i))
    block_list.append(int(i))
for i in range(num_sim):
    random.shuffle(napkin_list)
    for i in range(12):
        if napkin_list[i]==block_list[i]: #checks if value of napkin equals value of of block at same position
            fmatches+=1 #increases number of matches by 1 every time a block and napkin number correspond at same position for every iteration in a single simulation for a million simulations
print("Expected value: ", (fmatches/num_sim))#calculates mean as average of total number of matches observed over total number of simulations and prints results.

```

EXPLANATION

The Expected Value simulation is very similar to the Matching Napkins simulations. Instead of having a variable to store the number of simulations you were unlucky, you have a variable that stores the total number of matches per simulations for a large number of simulations. This gives the total number of matches, and you can find the average by dividing that number over the total number of simulations. Since we have a large number of simulations, our expected value gives us a number of matching napkins we should expect every time we have our random guess (which is 1). This value is independent of the number of napkins or blocks, even if there were 12 or 12000 napkins. This is because each napkin is assumed to have an equal chance of being picked (1/n), where n is the number of napkins in an independent event.

```python
from random import shuffle
import matplotlib.pyplot as plt #imports pyplot method from matplotlib library under variable plt
napkin_list=[]
block_list=[]
nummatches=[] #a list for storing the number of matches in each simulation 
num_sim=1000000
for i in range(1,13):
    napkin_list.append(int(i))
    block_list.append(int(i))
for i in range(num_sim):
    nmatches=0 #each time a new simulation is run sets nmatches varible to 0
    random.shuffle(napkin_list)
    for i in range(12):
        if napkin_list[i]==block_list[i]: #checks if value of napkin equals value of of block at same position
            nmatches+=1 #if so then increases the number of matches for that particular simulation every time condition is satisfied
    nummatches.append(nmatches) #at the end of 12 blocks for each simulation appends number of matches observed into nummatches list

#Histogram
fig,ax = plt.subplots() #function that returns a tuple containing a figure and axes objects which is unpacked into variables fig and ax
ax.hist(nummatches,label="Correctly matched napkins", edgecolor="black", alpha=0.7, bins=12) # accessor method that helps you set main components of histogram
ax.set(xlabel = 'Number of correctly matced napkins', ylabel = 'Frequency', title = 'Histogram showing frequency distributions for\n number of correctly matched napkins.') # accessor method that sets axes labels and title
plt.xticks(range(13)) #sets ticks on x-axis based on number of correctly matched napkins
plt.legend() #includes legend in histogram
plt.show() #dispalys histogram
```
